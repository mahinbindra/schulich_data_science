```python
from statsmodels.tsa.statespace.sarimax import SARIMAX
import itertools
import warnings

# Setting the range of parameters to test
p = d = q = range(0, 3)  # Considering 0, 1, 2 for non-seasonal parameters
pdq = list(itertools.product(p, d, q))  # All combinations of non-seasonal parameters
seasonal_pdq = [(x[0], x[1], x[2], 12) for x in pdq]  # Seasonal parameters with 12 months cycle
warnings.filterwarnings("ignore")  # Ignore warning messages
# Grid Search for parameter selection
best_aic = float("inf")
best_pdq = None
best_seasonal_pdq = None

for param in pdq:
    for param_seasonal in seasonal_pdq:
        try:
            mod = SARIMAX(monthly_aggregate,
                          order=param,
                          seasonal_order=param_seasonal,
                          enforce_stationarity=False,
                          enforce_invertibility=False)

            results = mod.fit(disp=0)
            if results.aic < best_aic:
                best_aic = results.aic
                best_pdq = param
                best_seasonal_pdq = param_seasonal
        except:
            continue

# Displaying the best parameters found
best_pdq, best_seasonal_pdq, best_aic
```

```python
# Calculate the average price for each SKU
average_prices = df.groupby('Product Code')['Value'].mean()

# Map the average prices to the 'Product Code' in the original DataFrame
df['Average Price'] = df['Product Code'].map(average_prices)

df.head()
```

```python
from sklearn.preprocessing import OneHotEncoder
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split

# Encoding the Product Code (SKU) using One-Hot Encoding
encoder = OneHotEncoder(sparse=False)
product_encoded = encoder.fit_transform(df[['Product Code']])

# Creating a Season variable based on the Order Date month
df['Season'] = df['Order Date'].dt.month % 12 // 3 + 1  # 1: Winter, 2: Spring, 3: Summer, 4: Fall

# Encoding the Season variable
season_encoded = encoder.fit_transform(df[['Season']])

# Preparing the dataset for logistic regression with both season and average price included
X = np.hstack((season_encoded, df[['Average Price']].values))
y = df['Product Code']  # Target variable

# Splitting the dataset into training and testing sets with the updated features
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.2, random_state=0
)

# Creating and training the logistic regression model with the new features
model_choice = LogisticRegression(multi_class='multinomial', solver='lbfgs')
model_choice.fit(X_train, y_train)
```

```python
from sklearn.metrics import classification_report

# Predicting on the test set
y_pred = model_choice.predict(X_test)

# Generating the classification report
report = classification_report(y_test, y_pred)

print(report)
```

```python
import numpy as np
import pandas as pd

# Number of simulations
num_simulations = 10000

# List to store results for each period
period_results_list = []

# List to store all simulated demands for confidence interval calculation
all_simulated_demands = np.zeros((num_simulations, 3))  # Assuming 3 SKUs

# Iterate over each period in the forecast_mean
for period_index, (period, aggregate_demand) in enumerate(forecast_mean.items()):
    # Running Monte Carlo simulations for each period
    period_sku_demands = np.zeros((num_simulations, 3))  # Reset for each period
    
    for sim in range(num_simulations):
        sku_demands = [0, 0, 0]  # Reset for each simulation
        current_probabilities = choice_probabilities[period_index]
        
        for _ in range(int(aggregate_demand)):
            chosen_sku_index = np.random.choice(len(current_probabilities), p=current_probabilities)
            sku_demands[chosen_sku_index] += 1

        # Record the demand for this simulation
        period_sku_demands[sim] = sku_demands

        # Aggregate across all simulations for confidence intervals
        all_simulated_demands[sim] += sku_demands

    # Calculating average demand per SKU for the period
    avg_sku_demand = period_sku_demands.mean(axis=0)
    # Append results for this period to the results list
    period_results_list.append([period] + avg_sku_demand.tolist())

# Create DataFrame from the period results list
columns = ['Period', 'SKU_A_Demand', 'SKU_B_Demand', 'SKU_C_Demand'] 
sku_demand_df = pd.DataFrame(period_results_list, columns=columns)

# Calculate overall mean and confidence intervals for each SKU
overall_mean_demand = all_simulated_demands.mean(axis=0)
confidence_intervals = np.percentile(all_simulated_demands, [2.5, 97.5], axis=0)

# Aggregate results with confidence intervals
aggregate_results = {
    'SKU': ['A', 'B', 'C'],
    'Mean Demand': overall_mean_demand,
    'Lower CI': confidence_intervals[0],
    'Upper CI': confidence_intervals[1]
}
aggregate_results_df = pd.DataFrame(aggregate_results)

# Displaying the DataFrames
print("Periodic Demand Forecasts:")
print(sku_demand_df)
print("\nOverall Demand Forecasts with Confidence Intervals:")
print(aggregate_results_df)
```
