```python
setosa = df[df['variety'] == 'Setosa']['sepal.width']
versicolor = df[df['variety'] == 'Versicolor']['sepal.width']
virginica = df[df['variety'] == 'Virginica']['sepal.width']
```

```python
from scipy import stats
```

```python
t_stat, p_val = stats.ttest_ind(setosa, versicolor)

print(f"T-statistic: {t_stat}")
print(f"P-value: {p_val}")
```

```python
t_stat, p_val = stats.ttest_ind(virginica, versicolor)

print(f"T-statistic: {t_stat}")
print(f"P-value: {p_val}")
```

```python
t_stat, p_val = stats.ttest_ind(virginica, setosa)

print(f"T-statistic: {t_stat}")
print(f"P-value: {p_val}")
```

```python
sns.lmplot(x = 'sepal.length', y = 'sepal.width', hue = 'variety', data = df)
```

```python
df['sepal_width_cat'] = pd.cut(df['sepal.width'], bins = [0,3,3.5,4])
contingency_table = pd.crosstab(df['variety'], df['sepal_width_cat'])
chi2, p_val, dof, expected = stats.chi2_contingency(contingency_table)

print(f"Chi-squared: {chi2}")
print(f"P-value: {p_val}")
```

```python
f_stat, p_val = stats.f_oneway(setosa, versicolor, virginica)

print(f"F-statistic: {f_stat}")
print(f"P-value: {p_val}")
```

```python
pearson_corr, p_val = stats.pearsonr(df['sepal.length'], df['sepal.width'])

print(f"Pearson's correlation: {pearson_corr}")
print(f"P-value: {p_val}")
```
